```python
import pandas as pd
```

```python
data1={'id':[1,2,3,4],'gender':['f','m','m','f']}
df1=pd.DataFrame(data1)
df1
```

```python
data2={'id':[1,2,3,4],'age':[25,30,22,38]}
df2=pd.DataFrame(data2)
df2
```

```python
pd.merge(df1,df2,left_on='id',right_on='id')
```

```python
data3={'id':[1,2,3,5],'job':['y','y','n','y']}
df3=pd.DataFrame(data3)
df3
```

```python
pd.merge(df1,df3,left_on='id',right_on='id')
```

```python
pd.merge(df1,df3,how='left',left_on='id',right_on='id')
```

```python
pd.merge(df1,df3,how='outer',left_on='id',right_on='id')
```

```python
pd.merge(df1,df3,how='right',right_on='id',left_on='id')
```

```python
pd.merge(df1,df3,how='inner',right_on='id',left_on='id')
```

```python
data4={'id':[1,2,3,5],'job':['YES','NO','YES','No']}
df4=pd.DataFrame(data4)
df4
```

```python
pd.merge(df3,df4,how='outer',right_on='id',left_on='id')
```

```python
pd.merge(df3,df4,how='outer',right_on='id',left_on='id',suffixes=('_left','_rigth'))
```
