```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airports=pd.read_csv('airports.csv')
```

```python
airports.loc[pd.isnull(airports['continent']),'continent']=='NA'
```

```python
airports['type'].value_counts().plot()
```

```python
airports['type'].value_counts().plot(kind='bar')
```

```python
airports['type'].value_counts().plot(kind='bar',figsize=(8,8))
plt.xlabel('airport_ type')
plt.ylabel('counter')
plt.show()
```

```python
airports['type'].value_counts().plot(kind='pie',figsize=(8,8))
plt.show()
```

```python
airports.groupby(['type','continent']).size()
```

```python
airports.groupby(['type','continent']).size().unstack()
```

```python
airports.groupby(['type','continent']).size().unstack().plot(kind='bar',figsize=(8,8))
plt.show()
```

```python
airports.groupby(['type','continent']).size().unstack().plot(kind='bar',figsize=(8,8),stacked=True)
plt.savefig('airports_type_continate-1.png')
plt.show()
```
