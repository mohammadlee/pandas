```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airports=pd.read_csv('airports.csv')
```

```python
airports['type'].value_counts().plot()
```

```python
airports['type'].value_counts().plot(kind='bar')
```

```python
airports['type'].value_counts().plot(kind='bar',figsize=(10,10))

```

```python
airports['type'].value_counts().plot(kind='bar',figsize=(10,10))
plt.xlabel('airports type')
plt.ylabel('count')
plt.show()
```

```python
airports['type'].value_counts().plot(kind='pie',figsize=(8,8))
plt.show()
```

```python
airports.groupby(['continent','type']).size()
```

```python
airports.groupby(['continent','type']).size().unstack()
```

```python
airports.groupby(['continent','type']).size().unstack().plot()
plt.show()
```

```python
airports.groupby(['continent','type']).size().unstack().plot(kind='bar',figsize=(8,8))
plt.show()
```

```python
airports.groupby(['continent','type']).size().unstack().plot(kind='bar',figsize=(8,8),stacked=True)
plt.savefig('airportss_type_containts.png')
plt.show()
```
